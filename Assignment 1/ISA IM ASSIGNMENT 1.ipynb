```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import math
```

```python
# For n = 7 which is the number of elements in our array
```

```python
mylist =[0.5,1,1.5,2,2.5,3,3.5]
```

```python
x = np.array(mylist)
```

```python
x
```

```python
# For Logarithmic = ùú≠((log n) plot
y1 = np.array([math.log(i,2) for i in x])
```

```python
y1
```

```python
# For Linear = ùú≠(n) plot
y2 = x
```

```python
y2
```

```python
# For Quadratic = ùú≠(n**2) plot
y3 = x**2
```

```python
y3
```

```python
# For Cubic = ùú≠(n**3) plot
y4 = x**3
```

```python
y4
```

```python
# For Polynomial = ùú≠(n**k) plot where k = 4
k=4
y5 = x**k
```

```python
y5
```

```python
# All equations in the same plot

fig = plt.figure(figsize = (8.5,8.5))

ax = fig.add_axes([0,0,1,1])
ax.plot(x,y1, label = r"$\dot{\Theta}$(log n)")
ax.plot(x,y2, label = r"$\dot{\Theta}$(n)")
ax.plot(x,y3, label = r"$\dot{\Theta}$(n**2)")
ax.plot(x,y4, label = r"$\dot{\Theta}$(n**3)")
ax.plot(x,y5, label = r"$\dot{\Theta}$(n**k) where k = 4")
plt.legend(bbox_to_anchor= (1.05,1), loc='upper left', fontsize='x-large')

ax.set_title(r"$\dot{\Theta}$(log n) VS $\dot{\Theta}$(n) VS $\dot{\Theta}$(n**2) VS $\dot{\Theta}$(n**3) VS $\dot{\Theta}$(n**k) where k = 4", fontsize = 'xx-large')
ax.set_xlabel('X')
ax.set_ylabel('Time')
```
